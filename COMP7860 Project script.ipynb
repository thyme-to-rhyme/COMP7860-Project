```
#cycles through the dataframe or list and inputs individual SMILES of compounds into script

topN=5
test_x=[]
test_px=[]

for ind in replicate.index: #df.index
  display(Chem.MolFromSmiles(replicate['smiles'][ind]))
  ringbreaker.show_topN_predictions(replicate['smiles'][ind], topN)
  print(ringbreaker.predict_ring_outcomes(replicate['smiles'][ind], topN))

#checks if there is anything in the dictionary indicating there is a prediction then appends the most probable prediction
  if ringbreaker.predict_ring_outcomes(replicate['smiles'][ind], topN)['prediction']:
    test_x.append(ringbreaker.predict_ring_outcomes(replicate['smiles'][ind], topN)['prediction'][0])
  else:
    test_x.append('NaN')

  if ringbreaker.predict_ring_outcomes(replicate['smiles'][ind], topN)['probability']:
    test_px.append(round(ringbreaker.predict_ring_outcomes(replicate['smiles'][ind], topN)['probability'][0],3))
  else:
    test_px.append('NaN')

replicate['Test x']=test_x
replicate['Test p(x)']=test_px
```
